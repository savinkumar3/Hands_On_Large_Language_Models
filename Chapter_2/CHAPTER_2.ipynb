```python
# Import Model snd Tokenizer from transformer Library
from transformers import AutoModelForCausalLM, AutoTokenizer
# Load model(Phi-3) and tokenizer
model = AutoModelForCausalLM.from_pretrained(
"microsoft/Phi-3-mini-4k-instruct",
# automatically choose the model CPU or GPU
device_map="auto",
# Choose the best data type automatically
torch_dtype="auto",
# Allow the model to run its own code
trust_remote_code=
True
,
)
# Load the tokenizer for the model
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
#Prompt given to the model to perform the task
# <|assistant|> tells the model that the  assistant need to  start generating a response
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened.<|assistant|>"
# Tokenizer convert  input prompt into tokens(numbers) so that model can understand
input_ids = tokenizer(prompt,
# PyTorch is a Python library used to build and train machine learning and deep learning models.
# returns the tokens as pytorch sensors
# Input_ids =Extract the token IDs
#.to("cpu") moves the tokens to the CPU
return_tensors="pt").input_ids.to("cpu")
# Generate the text
# Prints the token
for id in input_ids[0]:
  print(tokenizer.decode(id)) # Each token is converted back into text by using tokenizer.decode(id)
# Generate the response from the model
generation_output = model.generate(
input_ids=input_ids,  # given input tokens to the model prevously
max_new_tokens=20 # Genrated at most 20 new tokens
)
# Prints the response by converting token IDs back into normal text
print(tokenizer.decode(generation_output[0]))
# Decode the token ID 3323 to see which word it represents
print(tokenizer.decode(3323))
```

```python
#List of background colors
colors_list = [
'102;194;165', '252;141;98', '141;160;203',
'231;138;195', '166;216;84', '255;217;47'
]
# Function to display how  a tokenizer divide sentence  into tokens
def show_tokens(sentence, tokenizer_name):
  # Load the tokenizer from Hugging face model
  tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
  # Convert the sentence into token IDs
  token_ids = tokenizer(sentence).input_ids
  # loop for tokens
  # idx = Position of each token
  # t= token IDs
  for idx, t in enumerate(token_ids):
    # Print the each token with different color
    # decodes the tokens to the normal text
    print(f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m'  # idx % len(colors_list) cycles through the color list repeatedly.
          + tokenizer.decode(t) +
          '\x1b[0m', # \x1b[0m resets all text formatting (color, background color, bold, etc.) back to normal.
  end=' '
  )
```

```python
#Imports pandas library
# Pandas used to store and analyze the data
import pandas as pd
# Import request from urllib library
# It is used to download data from the URL
from urllib import request
# Get the playlist dataset file
#Downloads the playlist dataset from ther internet
# request.urlopen open the URL and return the file
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')
#Diplay the information about the downloaded file
print(data.info)
# Parse the playlist dataset file. Skip the first two lines as
# decode("utf-8") Converts the downloaded bytes into readdable  text
# split('\n')  Seperate the text  into individual lines
#  [2:] skips the first two lines because they contain metadata, not playlist data
lines = data.read().decode("utf-8").split('\n')[2:]
# Remove playlists with only one song
# rstrip() remove the extra space at the end of the line
#split() Seperate the song IDs based on space
#keep only the playlist that contain more than 1 song
playlists = [s.rstrip().split() for s in lines if len(s.split())
> 1]
# Load song metadata file
# A metadata file stores descriptive information about the main data.
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
# read  the metadata file and convert it into text
songs_file = songs_file.read().decode("utf-8").split('\n')
# Split the each line using tab character ('\t)
songs = [s.rstrip().split('\t') for s in songs_file]
# Create a Pandas DataFrame from the song metadata.
# The DataFrame will have three columns: id ,title,artist
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])
# set the id coloumn as index ,which may used for fast accessing
songs_df = songs_df.set_index('id')
```

```python
# print the first playlist and second playlist
print( 'Playlist #1:\n ', playlists[0], '\n')
print( 'Playlist #2:\n ', playlists[1])
```

```python
#Import word2vec model from gensim.models library
from gensim.models import Word2Vec
# Train our Word2Vec model
model = Word2Vec(
playlists,# Input data (previous playlist )
vector_size=32, # Each song is represented by a vector of size 32 number s
window=20,# Look at upto 20  neighbouring songs in the library
negative=50, # Uses 50 negative sample during training
min_count=1, # Include every song even it appears 1 time
workers=4 # uses 4 CPU cores to train the model faster
)
# select the song Id to which we want recommend
song_id = 2172
# Ask the model for songs similar to song  #2172
# positive=str(song_id) tells the model to search using this song by converting interger song Id to String because it stores the song Ids as String
model.wv.most_similar(positive=str(song_id))
```

```python
# Selects the song at row index 2172 from the songs_df DataFrame using iloc and prints all the details of that song on the screen.
print(songs_df.iloc[2172])
```

```python
#Import numpy library
# numpy is used to easu to calculate arrays
import numpy as np
#
def print_recommendations(song_id):
  # Finds the 5 songs that are similar to the given song
  # most_similar() returns a list of tuples:
  # (song_id, similarity_score)
  similar_songs = np.array(model.wv.most_similar(
      positive=str(song_id)# Convert song Id into string
      ,topn=5)
  )[:,0] # Extract only the song IDs
   # Use the recommended song IDs to retrieve the song title and artist
   # from the songs DataFrame.
return  songs_df.iloc[similar_songs]
# Display  recommendations by calling the function
print_recommendations(2172)
```
