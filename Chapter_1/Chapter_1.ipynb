```python
# Import Model and Tokenizer from the transformers library
from transformers import AutoModelForCausalLM, AutoTokenizer

# Loads the Phi-3 language model
model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-3-mini-4k-instruct",
     # Use the GPU
    device_map="cuda",
    # Choose the best data type automatically
    torch_dtype="auto",
     # Allow the model to run its own code
    trust_remote_code=True,
)

# Load the tokenizer for the model
tokenizer = AutoTokenizer.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct"
)
```

```python
# Import pipeline , which Combines the model and tokenizer  to generate the output
from transformers import pipeline
# Create a pipeline
generator = pipeline(
# Tells the pipeline to perform text generation task
"text-generation",
# Uses model that we are imported (Phi-3)
model=model,
# Uses the tokenizer that convert text into tokens and tokens to text =
tokenizer=tokenizer,
# return only newly generated text (Avoid the original prompt )
return_full_text=
False
,
# Allow the model to generate only 500 new tokens(words)
max_new_tokens=500,
# Tells the model to generate the specific(next) text (No randomness)
do_sample=
False
)
```

```python
# The prompt (user input / query)
messages = [{"role": "user", "content": "Create a funny joke about chickens.\n"}
]
# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```
